```python
cleaned_2020 = pd.read_csv('Final_Cleaned_2020EAVSData.csv')
cleaned_2016 = pd.read_csv('Final_Cleaned_2016EAVSData.csv')

connection_string = ######################################
client = MongoClient(connection_string)

db = client['eavs_db']

# Inserting data into collections
db.cleaned_2020.insert_many(cleaned_2020.to_dict('records'))
db.cleaned_2016.insert_many(cleaned_2016.to_dict('records'))

print("Data successfully loaded into MongoDB Atlas!")
```

### **Checking for Duplicates**


```python
cleaned_2020 = pd.read_csv('Final_Cleaned_2020EAVSData.csv')
cleaned_2016 = pd.read_csv('Final_Cleaned_2016EAVSData.csv')

# Function to check for duplicates and insert only new data
def insert_if_not_exists(collection_name, dataframe, unique_field):
    collection = db[collection_name]

    for record in dataframe.to_dict('records'):
        # Check if the record already exists
        if not collection.find_one({unique_field: record[unique_field]}):
            collection.insert_one(record)

    print(f"Finished processing {collection_name}.")

insert_if_not_exists('cleaned_2020', cleaned_2020, 'FIPSCode')
insert_if_not_exists('cleaned_2016', cleaned_2016, 'FIPSCode')

print("Data successfully checked and inserted into MongoDB!")
```

```python
import plotly.graph_objects as go
import plotly.express as px

#cleaned_2020 = pd.read_csv('Final_Cleaned_2020EAVSData.csv')
#cleaned_2016 = pd.read_csv('Final_Cleaned_2016EAVSData.csv')

cleaned_2020 = pandas_gbq.read_gbq("SELECT * FROM `ds2002final-443502.DS2002Final.2020EAVS`", project_id="ds2002final-443502")
cleaned_2016 = pandas_gbq.read_gbq("SELECT * FROM `ds2002final-443502.DS2002Final.2016EAVS`", project_id="ds2002final-443502")

cols_to_numeric_2020 = ['By-mail Transmitted Total', 'By-mail Returned By Voters Total', 'By-mail Counted Total', 'By-mail Rejected Total']

for i in cols_to_numeric_2020:
    cleaned_2020[i] = pd.to_numeric(cleaned_2020[i], errors='coerce')

bystate_2020 = cleaned_2020.groupby(['State_Abbr']).sum().reset_index()[['State_Abbr', 'By-mail Transmitted Total', 'By-mail Returned By Voters Total', 'By-mail Counted Total', 'By-mail Rejected Total']]

bystate_2020 = bystate_2020.drop([3, 8, 12, 27, 42, 50]).reset_index(drop = True).assign(Year = 2020) # Drops territories, adds year column

cols_to_numeric_2016 = ['By-mail Transmitted Total', 'By-mail Returned By Voters Total', 'By-mail Counted Total', 'By-mail Rejected Total']

for i in cols_to_numeric_2016:
    cleaned_2016[i] = pd.to_numeric(cleaned_2016[i], errors='coerce')

bystate_2016 = cleaned_2016.groupby(['State_Abbr']).sum().reset_index()[['State_Abbr', 'By-mail Transmitted Total', 'By-mail Returned By Voters Total', 'By-mail Counted Total', 'By-mail Rejected Total']]

bystate_2016 = bystate_2016.drop([7, 11, 40, 48]).reset_index(drop = True).assign(Year = 2016) # Drops territories, adds year column

bystate = pd.concat([bystate_2016, bystate_2020])
```

### Statistical Analysis

Question: Are the rates of mail-in ballot rejection by state different between 2016 and 2020?

To answer this, we run a one-way ANOVA test with year as our independent variable.

```python
import numpy as np
from scipy.stats import f_oneway
```

```python
bystate
```

```python
#Note: We have to add dropna() since Alabama does not report the number of By-mail ballots rejected. n=49 for this test.

f_oneway(bystate['Percent of By-mail ballots rejected'][bystate['Year'] == 2016].dropna(), bystate['Percent of By-mail ballots rejected'][bystate['Year'] == 2020].dropna())
```

We got a p-value of 0.0100, which is low enough to indicate that the distributions changed. But how did they change?

```python
fig = px.box(bystate, y='Percent of By-mail ballots rejected', x='Year', points = 'all')
fig.show()
```

From this, we see that though the highs and lows of the two distributions are similar, 2020 saw a large reduction in the percentage of mail-in ballots rejected across the states.
